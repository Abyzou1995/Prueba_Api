First task: Remove columns that will not be used, video,imdb_id,adult,original_title,vote_count,poster_path and homepage.

```python
df.drop(["video","imdb_id","adult","original_title","vote_count","poster_path", "homepage"],axis=1,inplace=True)
```

Second task: The null values ​​of the fields revenue, budget must be filled with the number 0

```python
df.budget.fillna(0,inplace=True)
df.revenue.fillna(0, inplace=True)
```

Third task: The null values from release_date field must be deleted.

```python
df.dropna(subset=['release_date'],inplace=True)
df.reset_index(drop=True)
df
```

Third task :If there are dates, they must have the format YYYY-mm-dd, and it must also create the release_year column where they will extract the year of the release date.

```python
df.release_date=pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')

```

```python
df["release_year"]=df.release_date.dt.year
df.reset_index(drop=True,inplace=True)
df.dropna(subset=['release_date'], inplace=True)
df['release_year'] = df['release_year'].astype(int, errors = 'raise')
```

Fourth task: The null values from release_date field must be deleted.

```python
df.dropna(subset=['release_date'], inplace=True)
df.reset_index(drop=True,inplace=True)

```

Fifth task: Create the column with the return on investment, called return with the revenue and budget fields, dividing these last two revenue / budget, when there is no data available to calculate it, it must take the value 0.

```python
df['revenue'] = df['revenue'].astype(float, errors = 'raise')
df['budget'] = df['budget'].astype(float, errors='raise')
```

```python
df["return"]=df.revenue.div(df.budget.replace(0,np.nan)).fillna(0)
```

Sixth Task:Some fields, like belongs_to_collection, production_companies and others (see data dictionary) are nested, that is, they either have a dictionary or a list as values ​​in each row, you will need to un-nest them to be able to join the dataset again to do any of the queries! from the API! Or find a way to access that data without un-nested.

Belongs_to_collection => Dictionary

```python
def convert_diccionario(diccionario):
    if isinstance(diccionario,str) and "{" in diccionario:
        dic=ast.literal_eval(diccionario)
        return dic["name"]
```

```python
df["belongs_to_collection"]=df["belongs_to_collection"].apply(convert_diccionario)
```

Production_companies => List

```python
def convert_lista(lista):
    if isinstance(lista, str) and "{" in lista:
        L=[]
        for i in ast.literal_eval(lista):
            L.append(i["name"])
        return L
```

```python
df["production_companies"]=df["production_companies"].apply(convert_lista)
```

```python
df["genres"]=df["genres"].apply(convert_lista)
df["production_countries"]=df["production_countries"].apply(convert_lista)
##df["spoken_languages"]=df["spoken_languages"].apply(convert_lista)
```
